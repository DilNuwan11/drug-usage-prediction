```python
import pandas as pd
```

## Cleaning drug usage crime data

```python
df = pd.read_csv('./raw/Reported_drug_crimes_by_regions.csv', encoding='utf-8')
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
df_sorted_columns = df.reindex(sorted(df.columns), axis=1)
df_sorted_columns.head()
```

```python
df_regions = df_sorted_columns.drop(columns=['MA1 MANNER-SUOMI', 'MA2 AHVENANMAA'])
df_regions.head()
```

```python
df_regions.columns
```

```python
# Deleting the region code
df_regions.columns = ['KOKO MAA'] + [col.split()[1] for col in df_regions.columns if col[0] == 'M'] + ['year']
df_regions.head()
```

```python
df_regions.to_csv('./clean/Reported_drug_usage_by_regions.csv', encoding='utf-8')
```

### Exporting regions' names

```python
regions = df_regions.columns

with open('region_names.txt', 'w', encoding='utf-8') as f:
    for region in regions:
        if (region != 'year') and (region != 'KOKO MAA'):
            f.write(f"{region}\n")
```

## Cleaning Rehabilitated Patients Data

```python
data = pd.read_csv('raw/Rehabilitation.csv')
```

```python
data = data.iloc[:, 0].str.split(';', expand=True)
```

```python
print(data.head())
```

```python
data = data.drop(columns=[0,1,3,4,7])
```

```python
print(data.head())
```

```python
data.columns = ['Area', 'Year','no_of_patients_rehabilitated']
print(data.head())
```

```python
data_pivot = data.pivot(index='Year', columns='Area', values='no_of_patients_rehabilitated').reset_index()

data_pivot.columns = ['Year', 'Central Finland', 'Central Ostrobothnia', 'Kainuu', 'Kanta-Häme', 'Kymenlaakso', 'Lapland', 'North Karelia', 'North Ostrobothnia', 'North Savo', 'Ostrobothnia', 'Pirkanmaa', 'Päijät-Häme', 'Satakunta', 'South Karelia', 'South Ostrobothnia', 'South Savo', 'Southwest Finland', 'Uusimaa', 'Åland']

data_pivot.head()
```

```python
data_pivot.columns = ['Year', 'Keski-Suomi', 'Keski-Pohjanmaa', 'Kainuu', 'Kanta-Häme', 'Kymenlaakso', 'Lappi', 'Pohjois-Karjala', 'Pohjois-Pohjanmaa', 'Pohjois-Savo', 'Pohjanmaa', 'Pirkanmaa', 'Päijät-Häme', 'Satakunta', 'Etelä-Karjala', 'Etelä-Pohjanmaa', 'Etelä-Savo', 'Varsinais-Suomi', 'Uusimaa', 'Ahvenanmaa']
print(data_pivot.head())
```

```python
print(data_pivot.shape)
print(data_pivot.describe())

data_pivot = data_pivot.reindex(sorted(data_pivot.columns[1:]), axis=1)

data_pivot.to_csv('clean/Rehabilitation_data_by_regions.csv', index=False)
```
