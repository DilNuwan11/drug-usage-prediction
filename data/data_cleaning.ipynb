## Cleaning drug usage crime data

```python
df = pd.read_csv('./raw/Reported_drug_crimes_by_regions.csv', encoding='utf-8')
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
df_sorted_columns = df.reindex(sorted(df.columns), axis=1)
df_sorted_columns.head()
```

```python
df_regions = df_sorted_columns.drop(columns=['MA1 MANNER-SUOMI', 'MA2 AHVENANMAA'])
df_regions.head()
```

```python
df_regions.columns
```

```python
# Deleting the region code
df_regions.columns = ['KOKO MAA'] + [col.split()[1] for col in df_regions.columns if col[0] == 'M'] + ['year']
df_regions.head()
```

```python
df_regions.to_csv('./clean/Reported_drug_usage_by_regions.csv', encoding='utf-8')
```

### Exporting regions' names

```python
regions = df_regions.columns

with open('region_names.txt', 'w', encoding='utf-8') as f:
    for region in regions:
        if (region != 'year') and (region != 'KOKO MAA'):
            f.write(f"{region}\n")
```

## Clean Unemployment rate

```python
df = pd.read_csv('./raw/Unemployment_rate.csv', encoding='utf-8')
df.head()
```

```python
# Deleting the region code
df_regions = df
df_regions.columns = ['Year', 'WHOLE COUNTRY'] + [col.split(' ', 1)[1] for col in df.columns if col.startswith('MK')]
df_regions.head()
```

```python
# Make sure the region names match
sorted(df_regions.drop(columns=['WHOLE COUNTRY']).columns) == sorted(data_pivot.columns)
```

### imputing missing data

```python
df_regions.describe()
```

```python
import numpy as np

df_regions.replace('.', np.nan, inplace=True)
df_regions.head()
```

```python
nan_columns = df_regions.columns[df_regions.isna().any()].tolist()
print(nan_columns)
```

```python
for col in nan_columns:
    print(df_regions[col])
    
```

#### Kainuu

```python
df_regions['Kainuu'] = pd.to_numeric(df_regions['Kainuu'], errors='coerce')
df_regions['Kainuu'].plot()
df_regions['WHOLE COUNTRY'].plot()

import matplotlib.pyplot as plt

plt.legend(['Kainuu', 'WHOLE COUNTRY'])
plt.show()

# I will impute just the last years with whole country data.
```

```python
df_regions['Kainuu'] = df_regions['Kainuu'].fillna(df_regions['WHOLE COUNTRY'])
df_regions['Kainuu'].plot()
df_regions['WHOLE COUNTRY'].plot()

plt.legend(['Kainuu', 'WHOLE COUNTRY'])
plt.show()
```

#### Central Ostrobothnia

```python
df_regions['Central Ostrobothnia'] = pd.to_numeric(df_regions['Central Ostrobothnia'], errors='coerce')
df_regions['Central Ostrobothnia'].plot()
df_regions['WHOLE COUNTRY'].plot()

plt.legend(['Central Ostrobothnia', 'WHOLE COUNTRY'])

# I will impute the entire column with whole country data.
```

```python
df_regions[['Central Ostrobothnia', 'Åland']] = df_regions[['Central Ostrobothnia', 'Åland']].apply(lambda x: x.fillna(df_regions['WHOLE COUNTRY']))
df_regions.head()
```

```python
df_regions.to_csv('./clean/Unemployment_rate_by_regions.csv', encoding='utf-8')
```

### melting the data

```python
def melting_data(df, value_name):
    df_melted = pd.melt(df_regions, id_vars=['Year'], 
                    value_vars= ['Uusimaa', 'Southwest Finland', 'Satakunta',
                                'Kanta-Häme', 'Pirkanmaa', 'Päijät-Häme', 'Kymenlaakso',
                                'South Karelia', 'South Savo', 'North Savo', 'North Karelia',
                                'Central Finland', 'South Ostrobothnia', 'Ostrobothnia',
                                'Central Ostrobothnia', 'North Ostrobothnia', 'Kainuu', 'Lapland',
                                'Åland'],
                    var_name='region', value_name=value_name)
    return df_melted

df_melted = melting_data(df_regions, 'unemployment_rate')
df_melted
```

```python
df_melted.to_csv('./clean/melted/Unemployment_rate.csv', encoding='utf-8')
```
